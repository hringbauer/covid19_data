## Plot some Bessel Functinos

```python
xlim =[0,20]
x = np.linspace(0.1,20, 10000)

y0 = kv(0, x)
y1 = x * kv(1, x)
y2 = x**2 * kv(2, x)
y3 = x**3 * kv(3, x)

cmap = cm.get_cmap('viridis_r')
cs = cmap(np.linspace(0,0.6,4))
```

```python
savepath=("./figures/dumpster/pw_fraction.png")
c=0.01

plt.figure(figsize=(4,4))
ax = plt.gca()
ax.plot(x, c*y0, label=f"$0$", lw=2, c=cs[0])
ax.plot(x, c*y1, label=f"$1$", lw=2, c=cs[1])
ax.plot(x, c*y2, label=f"$2$", lw=2, c=cs[2])
ax.plot(x, c*y3, label=f"$3$", lw=2, c=cs[3])
ax.legend(title="Mutations apart")
ax.set_xlim(xlim)
ax.set_yscale("log")
ax.set_xlabel(f"Pairwise geographic Distance [$\sigma$]")
ax.set_ylabel("Fraction of sample pairs")
ax.set_title("")

if len(savepath)>0:
    plt.savefig(savepath, bbox_inches = 'tight', pad_inches = 0, dpi=400)
    print(f"Saved figure to {savepath}")

plt.show()
```
